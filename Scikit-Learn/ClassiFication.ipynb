**Logistic Regression**

1->It is a classification Algorithm
2->Output is between 0-1
3->Used to predict the Binary Output(yes or No)


```python
from sklearn.linear_model import LogisticRegression

X=[[1],[2],[3],[4],[5]] #input->Hours Studies
y=[0,0,1,1,1]


model=LogisticRegression()

model.fit(X,y)

hours=float(input("Enter How many Hours tou Studied: "))

result = model.predict([[hours]])[0] #0->to get output in 1 and 0 format

if result==1:
    print(f"Based on the hours {hours} you studied you are likely to PASS")
else:
    print(f"Based on the hours {hours} you studied you are likely to FAIL")
    
    
```

**KNN-k nearest Neighbour**

Steps:
1->Choose K(neighBours Number)
2->K(Closet datapoint from the Neoghbours)
3->Calc How many are in niche class(Y or N)

```python
from sklearn.neighbors import KNeighborsClassifier
X=[
    [180,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]
]

#0->Apple & 1->orange

y=[0,0,0,1,1,1]

model=KNeighborsClassifier(n_neighbors=3)

model.fit(X,y)

weight=float(input("Enter The weight of  fruit in grams: "))
size=float(input("Enter The size of  fruit: "))

predicted=model.predict([[weight,size]])[0]


if predicted==0:
    print("It is likely to be an Apple")
else:
    print("It is likely to be orange")
```

**Decision Tree**


```python
from sklear.tree import DecisionTreeClassifier

X=[
    [7,2],#Apple
    [8,3],#Apple
    [9,8],#orange
    [10,9]#orange
]

y=[0,0,1,1] #0->Apple,1->Orange

model=DecisionTreeClassifier()

model.fit(X,y)

size=float(input("Enter The size of  fruit in cm: "))
shade=int(input("Enter The shade of fruit(0-10): "))

int result=model.predict([[size,shade]])[0]

if result==0:
    print("This is likely to be apple')
else:
    print("This is likely to be Orange")
    


```
