```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weaks_data=[1,2,3,4,5]
sales_data=[2,4,5,4,5]

df=pd.DataFrame({"Weeks":weaks_data,
                 "Sales":sales_data})
df
```

```python
X=df["Weeks"].values
Y=df["Sales"].values
print(X)
print(Y)
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)

n=len(X)

numer=0
denom=0 #Counters

for i in range(n):
    numer+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom+=(X[i]-mean_x)**2

m=numer/denom
c=mean_y-(m*mean_x)

print(f"m:{m}, c:{c}")
    
```

```python
#Plot
x_max=np.max(X)+1
x_min=np.min(X)-1

x=np.linspace(x_min,x_max)
y=(m * x)+c


plt.plot(x,y,color="#8c250f",label="Regression Line",linestyle="--")
plt.scatter(X,Y,color="red",label="Actual Data")
plt.xlabel("Weeks")
plt.ylabel("Sales")
plt.title("LR Demo")
plt.grid(True,alpha=0.2,linestyle=":")
plt.legend(loc="best")
plt.show()
```

```python
#Calculate r2 vlaue

ss_tot = 0
ss_res = 0

for i in range(n):
    y_pred = (m * X[i] + c) # Use addition for y_pred calculation
    ss_tot += (Y[i] - mean_y)**2 # Accumulate the sum
    ss_res += (Y[i] - y_pred)**2 # Accumulate the sum

r2 = 1 - (ss_res / ss_tot) # Calculate R2 after the loop
print(f"R2 is: {r2}")
```

```python
#Validation

from sklearn.linear_model import LinearRegression
X=X.reshape((n,1))
```

```python
reg=LinearRegression()
reg=reg.fit(X,Y)
```

```python
R2=reg.score(X,Y)
R2
```
