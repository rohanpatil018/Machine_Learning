```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x=np.array([])
y=np.array([10,12,16,14,18,22,22,20])
plt.plot(y)
plt.show()
```

```python
days=['Mon','Tue','Wed','Thur','Fri']
sales=[7,10,5,20,11]


plt.title("Bakery Report",c='g')
plt.xlabel("Days")
plt.ylabel("Sales")
plt.plot(days,sales,marker='o')
plt.show()
```

```python
days=['Mon','Tue','Wed','Thur','Fri']
sales=[7,10,5,20,11]
plt.bar(days,sales,color='b',label='Sales in a Month')
# plt.show()
plt.legend()
plt.show()
```

```python
region=['North','South','East','West']
revenue=[2000,1600,1900,1200]
label=['N','S','E','W']
color=['Red','Green','Blue','Orange']

plt.title("Region Wise Revenue")
plt.pie(revenue, labels=label,colors=color,autopct=lambda p: f'${(p*sum(revenue)/100):.0f}')
plt.legend()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,5,2,9,8]

fig,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].plot(x,y,color='b',marker='h')
ax[0].set_title("Line Chart")


ax[1].bar(x,y,color='r')
ax[1].set_title("Bar Chart")

plt.tight_layout()

plt.show()
```
